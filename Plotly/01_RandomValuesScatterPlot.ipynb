## Program to draw a scatter plot for random 1000 x and y coordinates

```python
import plotly.express as px
import numpy as np

# creating arrays for x-axis values & y-axis values
x_values = np.random.randn(1000)
y_values = np.random.randn(1000)

# constructing scatter plot 
fig = px.scatter(x=x_values, y=y_values)
fig                   # displaying plot
```
