##  Program to draw a scatter plot for random 500 x and y coordinates and style it


```python
import plotly.graph_objects as go
import numpy as np

# creating arrays for x-axis values & y-axis values
x_values = np.random.randn(500)
y_values = np.random.randn(500)

# constructing line & scatter plots and styling marker using dictionary as argument
fig = go.Figure(go.Scatter(x=x_values, y=y_values, mode='markers', marker=dict(size=12, line=dict(width=2,
                                        color='DarkSlateGray' ))))
fig.update_layout(title='Styled Scatter Plot')      # giving title to graph
fig                   # displaying plots
```
