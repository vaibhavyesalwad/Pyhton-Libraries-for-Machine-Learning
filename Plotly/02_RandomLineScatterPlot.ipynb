## Program to draw line and scatter plots for random 100 x and y coordinates

```python
import plotly.graph_objects as go
import numpy as np

# creating arrays for x-axis values & y-axis values
x_values = np.random.randn(100)
y_values = np.random.randn(100)

# constructing line & scatter plots 
fig = go.Figure(go.Scatter(x=x_values, y=y_values, mode='markers+lines'))
fig                   # displaying plots
```
