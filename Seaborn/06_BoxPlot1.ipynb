## Program to draw box plot of life expectancy by continent for a data set given in a url 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# loading dataset from file
gapminder = pd.read_csv('gapminder-FiveYearData.csv')
# plotting violin plot 
sns.boxplot(x='continent', y='lifeExp', data=gapminder)
plt.show()               # drawing box plot
```
