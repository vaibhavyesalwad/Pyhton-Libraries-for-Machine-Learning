## Program to draw a point plot for sex against survived for a dataset given in url


```python
import seaborn as sns
import matplotlib.pyplot as plt
# loading dataset from file
titanic = sns.load_dataset('titanic')
titanic
```

```python
# constructing point plot for above dataset
sns.pointplot(x='sex',y='survived',data=titanic)
plt.show()                 # drawing point-plot
```
