## Program to draw a scatter plot of “day” against “total bill” for a dataset given in a url

```python
import seaborn as sns
import matplotlib.pyplot as plt
# loading dataset from file
tips = sns.load_dataset('tips')
# plotting scatter plot day vs total_bill
sns.relplot(x='day',y='total_bill', kind='scatter', data=tips)
plt.show()           # drawing graph
```
