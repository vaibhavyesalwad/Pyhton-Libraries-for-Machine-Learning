## Program to draw a line with suitable label in the x axis, y axis and a title

```python
import matplotlib.pyplot as plt
x = range(10)                     # creating list of values 
plt.plot(x,x,label=('y = x'))     # plotting graph 'x' vs 'y' (here y = x)
plt.xlabel('x-axis')              # labeling list of values on x-axis
plt.ylabel('y-axis')              # labeling list of values on y-axis
plt.title('First Graph')          # giving title to graph
plt.legend()                      # box naming lines or naming plotted points 
plt.draw()                        # drawing current graph- alternate way plt.show()
```
