## Program to plot several lines with different format styles in one command using arrays

```python
import matplotlib.pyplot as plt
import numpy as np
# creating collection of values
numbers = np.arange(0,5,.2)
plt.plot(numbers, numbers, 'r-', numbers, numbers**2, 'b-', numbers, numbers**3, 'g^')   # plotting 
plt.xlabel('numbers')
plt.ylabel('exponents')         # labeling axes
plt.show()                      # drawing graph
```
