## Program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with linestyle -, width .5. and color blue

```python
import matplotlib.pyplot as plt
# Storing given data in list
data = [
        ["03-10-16",772.559998],
        ["04-10-16",776.429993],
        ["05-10-16",776.469971],
        ["06-10-16",776.859985],
        ["07-10-16",775.080017]
      ]
# plotting graph date vs closing values
plt.plot([row[0] for row in data], [row[1] for row in data],label= 'Date vs Alphabet Inc Closing values')
plt.grid(ls='-', lw=0.5, color ='blue')                # customising grid lines 
plt.title("Closing values of Alphabet Inc stocks")     # giving title
plt.xlabel('Date')                                      
plt.ylabel('Closing values')                           # lebeling axes 
plt.legend()                                           # box - naming graph
plt.show()                                             # drawing graph
```
