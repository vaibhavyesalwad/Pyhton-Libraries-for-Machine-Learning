## Program to draw a scatter graph taking a random distribution in X and Y and plotted against each other

```python
import matplotlib.pyplot as plt
import numpy as np
# generating random values for two numpy arrays
numbers1 = np.random.randn(80)
numbers2 = np.random.randn(80)
# plotting scatterplot
plt.scatter(numbers1, numbers2)
plt.title('Scatterplot for random values on x-axis & y-axis')      # defining title
plt.xlabel('x')
plt.ylabel('y')                                                    # labeling axes
plt.show()                                                         # draw graph
```
