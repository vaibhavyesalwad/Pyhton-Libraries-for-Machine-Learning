## Program to draw a scatter plot using random distributions to generate balls of different sizes

```python
import matplotlib.pyplot as plt
import numpy as np
# generating random values for two numpy arrays
numbers1 = np.random.randn(80)
numbers2 = np.random.randn(80)
sizes = np.random.randint(1,80,size=80) 
# plotting scatterplot with different size circles
plt.scatter(numbers1, numbers2, s=sizes, facecolor ='none', edgecolor='r')
plt.title('Scatterplot for random values on x-axis & y-axis')      # defining title
plt.xlabel('x')
plt.ylabel('y')                                                    # labeling axes
plt.show()                                                         # draw graph
```
