## Program to display a horizontal bar chart of the popularity of programming Languages

```python
import matplotlib.pyplot as plt
# values for y-axis 
prog_langs = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
# bar heights for values on y-axis
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
# plotting horizontal bar chart
plt.barh(prog_langs,popularity)
plt.ylabel('Programming Languages')      # labeling y-axis
plt.xlabel('Popularity index')                 # labeling x-axis
plt.show()                               # drawing bar chart
```
