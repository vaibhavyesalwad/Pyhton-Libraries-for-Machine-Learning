## Program to create bar plots with errorbars on the same figure. Attach a text label above each bar displaying men means (integer value)

```python
import matplotlib.pyplot as plt
import numpy as np
# defining data for bars
means = [0.2474, 0.1235, 0.1737, 0.1824]
std_deviations = [0.3314, 0.2278, 0.2836, 0.2645]     # it is used as error
# defining positions of bars
positions = np.arange(0,2*len(means),2) 
mygraph = plt.bar(positions, means, yerr = std_deviations,capsize =10)     # plotting bar graph with error bars
# defining xticks at given positions
samples =['sample1','sample2','sample3','sample4']
plt.xticks(positions, samples)
# printing bar values above each bar
for mybar in mygraph.patches:
    plt.text(mybar.get_x()+.5, mybar.get_height()+0.02, mybar.get_height())
plt.title('Bar plots with error bars')       # defining title
plt.ylabel('means of velocities')               
plt.xlabel('velocity')                       # labels for axes
plt.show()                                   # drawing plot
```
