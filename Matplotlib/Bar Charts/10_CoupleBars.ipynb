##  Program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)


```python
import matplotlib.pyplot as plt
import numpy as np

men = [20,30,35,35,26]
women = [25,32,30,35,26]

# defining positions for each bar for men values
positions=np.arange(1,3*len(men),3)

# plotting bar for men values at defined positions
plt.bar(positions, men, label='men',width=1)
# plotting bar for women values at beside bar of men values
plt.bar(positions+1, women, label='women',width=1)

group_labels = ['Group1','Group2','Group3','Group4','Group5']
# writing ticks on x-axis
plt.xticks(positions+.5,group_labels)
plt.title('Mean values for men & women in different groups')
plt.xlabel('Groups comprising men & women')           # labeling x-axis
plt.ylabel('Means')            # labeling y-axis
plt.legend()                   # box - naming bars
plt.show()                     # drawing bar chart
```
