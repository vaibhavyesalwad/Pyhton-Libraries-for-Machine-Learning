## Program to create bar plots with error bars on tht same figure.

```python
import matplotlib.pyplot as plt
import numpy as np
# defining data for bars
means = [0.2474, 0.1235, 0.1737, 0.1824]
std_deviations = [0.3314, 0.2278, 0.2836, 0.2645]
# defining positions of bars
positions = np.arange(0,2*len(means),2) 
plt.bar(positions, means, yerr = std_deviations,capsize =10)     # plotting bar graph with error bars
# defining xticks at given positions
samples =['sample1','sample2','sample3','sample4']
plt.xticks(positions, samples)
plt.title('Bar plots with error bars')       # defining title
plt.ylabel('means of velocities')               
plt.xlabel('velocity')                       # labels for axes
plt.show()                                   # drawing plot
```
