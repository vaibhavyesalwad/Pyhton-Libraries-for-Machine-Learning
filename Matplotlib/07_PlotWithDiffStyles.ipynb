## Program to plot two or more lines with different styles

```python
import matplotlib.pyplot as plt
import numpy as np
numbers = np.linspace(0,3,100)                     # creating list of values 
plt.style.use('ggplot')                           # using 'ggplot' style
plt.plot(numbers,numbers**2,label=('y = x'))     # plotting graph 'x' vs 'y' (here y = x)
plt.xlabel('x-axis')              # labeling list of values on x-axis
plt.ylabel('y-axis')              # labeling list of values on y-axis
plt.title("Using 'ggplot' style")                  # giving title to graph
plt.legend()                      # box naming lines or naming plotted points 
plt.draw()     
```

```python
plt.style.use('grayscale')                        # using 'grayscale' style
plt.plot(numbers,numbers**2,label=('y = x'))     # plotting graph 'x' vs 'y' (here y = x)
plt.xlabel('x-axis')              # labeling list of values on x-axis
plt.ylabel('y-axis')              # labeling list of values on y-axis
plt.title("Using 'grayscale' style")          # giving title to graph
plt.legend()                      # box naming lines or naming plotted points 
plt.draw()                        # drawing current graph- alternate way plt.show()
```

```python
plt.style.use('classic')                        # using 'scatterplot' style
plt.plot(numbers,numbers**2,label=('y = x'))     # plotting graph 'x' vs 'y' (here y = x)
plt.xlabel('x-axis')              # labeling list of values on x-axis
plt.ylabel('y-axis')              # labeling list of values on y-axis
plt.title('')          # giving title to graph
plt.legend()                      # box naming lines or naming plotted points 
plt.draw()     
```

```python
plt.style.use('bmh')                        # using 'scatterplot' style
plt.plot(numbers,numbers**2,label=('y = x'))     # plotting graph 'x' vs 'y' (here y = x)
plt.xlabel('x-axis')              # labeling list of values on x-axis
plt.ylabel('y-axis')              # labeling list of values on y-axis
plt.title('')          # giving title to graph
plt.legend()                      # box naming lines or naming plotted points 
plt.draw()     
```
