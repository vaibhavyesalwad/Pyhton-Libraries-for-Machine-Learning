## Program to create a pie chart with a title of the popularity of programming Languages

```python
import matplotlib.pyplot as plt
# labels for slices in pie chart
prog_langs = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
# value for each slice of pie chart
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
# plotting pie chart using values in data of popuplarity & adding labels to slices
plt.pie(popularity, labels= prog_langs, autopct='%1.2f%%')
plt.title('Popularity of Programming Languages')  # added title
plt.show()                                # drawing pie chart
```
