## Program to plot two or more lines on same plot with suitable legends of each line

```python
import matplotlib.pyplot as plt
import numpy as np
nums = np.linspace(0,3,100)
plt.plot(nums,nums,label = 'y = x')
plt.plot(nums,nums**2,label = 'y = x**2')
plt.plot(nums,nums**3,label = 'y = x**3')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plotting 2 or more line graphs on same plot')
plt.legend()
plt.show()
```
