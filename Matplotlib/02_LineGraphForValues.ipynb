## Program to draw a line using given axis values with suitable label in the x axis , y axis and a title

```python
import matplotlib.pyplot as plt
heights = [165,167,168,172,177,182,186,189,190]          # creating list of values       
weights = [57,59,64,69,75,83,90,97,100]               # corresponding values for above list of values
plt.plot(heights,weights, label= 'heights vs weights')     # plotting graph 'x' vs 'y' 
plt.xlabel('heights(cm)')              # labeling list of values on x-axis
plt.ylabel('weights(kg)')              # labeling list of values on y-axis
plt.title('Relation between heights & weights')          # giving title to graph
plt.legend()                      # box- naming lines or naming plotted points 
plt.draw()                        # drawing current graph- alternate way plt.show()
```
