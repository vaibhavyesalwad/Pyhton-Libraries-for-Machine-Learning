## Program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title

```python
import matplotlib.pyplot as plt
# reading lines from text file
with open('values.txt','r') as file:
    data = file.readlines()

# extracting values for x-axis
x_values = [int(line.split()[0]) for line in data]
x_values
```

```python
# extracting values for y-axis
y_values = [int(line.split()[1]) for line in data]
y_values
```

```python
plt.plot(x_values,y_values)     # plotting graph for given lists of values
plt.xlabel('x_values')          # labeling x-axis
plt.ylabel('y_values')          # labeling y-axis
plt.show()                      # showing graph
```
