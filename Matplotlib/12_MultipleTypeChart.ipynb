##  Program to create multiple types of charts

```python
import matplotlib.pyplot as plt
import numpy as np
# using numpy.arange for creating array of values
numbers = np.arange(10)
# plotting graph for squares
plt.plot(numbers,numbers*2,'rs', numbers, numbers*2, 'r-')  # plotting different charts for same values 
plt.title('')
plt.xlabel('numbers') 
plt.ylabel('squares')      # labeling axes
plt.show()                 # drawing graph
```
