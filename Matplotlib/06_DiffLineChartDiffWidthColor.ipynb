## Program to plot two or more lines with legends, different widths and colors

```python
import matplotlib.pyplot as plt
import numpy as np
# using numpy.linspace function gives evenly spaced numbers
numbers = np.linspace(0,3,100)
# color & linewidth parameters for the line graph 
plt.plot(numbers, numbers**2, label = 'y = x**2',color = 'blue', linewidth= 2)
plt.plot(numbers, numbers**3, label = 'y = x**3',color ='red', linewidth = 4)
plt.xlabel('numbers')           # labeling x-axis
plt.ylabel('exponents')         # labeling y-axis
plt.legend()                    # box - naming different plots
plt.draw()                      # drawing graph   
```
