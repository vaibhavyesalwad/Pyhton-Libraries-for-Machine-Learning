## Program to display the current axis limits values and set new axis values

```python
import matplotlib.pyplot as plt
import numpy as np
# using numpy.linspace function for evenly spreaded values 
nums = np.linspace(0,4,50)
plt.plot(nums,3*nums+5,label='y = 3x+5')    # plotting graph
# printing current axes limits 
print(f'Current x-axis limits: {plt.xlim()}')
print(f'Current y-axis limits: {plt.ylim()}')
plt.xlabel('x')
plt.ylabel('y')          # labeling axes
plt.legend()             # box - naming graph
plt.show()               # drawing graph
```

```python
plt.plot(nums,3*nums+5,label= 'y = 3x+5')  # plotting graph
# changing current x-axis limits
plt.xlim(-2,6)          # can be done using keyword arguments plt.xlim(left= ,right= )
# changing current y-axis limits
plt.ylim(0,20)          # keyword arguments plt.ylim(bottom= ,top= )
plt.xlabel('x')          
plt.ylabel('y')          # labeling axes
plt.legend()             # box - naming graph
plt.show()               # drawing graph
```
