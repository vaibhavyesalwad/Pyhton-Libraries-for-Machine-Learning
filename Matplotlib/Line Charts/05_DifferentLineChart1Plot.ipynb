## Program to plot two or more lines on same plot with suitable legends of each line

```python
import matplotlib.pyplot as plt
import numpy as np
# using numpy.linspace() for evenly distributed values 
numbers = np.linspace(0,3,100) 
# plotting 3 graphs on same plot
plt.plot(numbers,numbers,label = 'y = x')
plt.plot(numbers,numbers**2,label = 'y = x**2')
plt.plot(numbers,numbers**3,label = 'y = x**3')
plt.xlabel('numbers')              # labeling x-axis 
plt.ylabel('exponents')
plt.title('Plotting 3 line graphs on same plot')
plt.legend()                      # box -naming different line graphs
plt.show()                        # drawing graphs
```
