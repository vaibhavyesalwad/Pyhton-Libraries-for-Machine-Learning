## Program to create multiple plots

```python
import matplotlib.pyplot as plt 
import numpy as np
# creating numpy array of values 
numbers = np.linspace(-3,3,100)
# initialiazing fig
fig = plt.figure()

# defining subplots
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

# plotting subplots
ax1.plot(numbers, numbers**2)
ax2.plot(numbers, numbers**3)
ax3.plot(numbers, numbers**4)
ax4.plot(numbers, numbers**5)
fig.suptitle("Squares, Cubes, Quadraples, number**5 of [-3,3]")
plt.show()
```
