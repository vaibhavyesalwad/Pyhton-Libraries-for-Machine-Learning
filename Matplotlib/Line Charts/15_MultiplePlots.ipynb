## Program to create multiple plots

```python
import matplotlib.pyplot as plt 
import numpy as np
# creating numpy array of values 
numbers = np.linspace(-5,5,100)
# initialiazing fig
fig = plt.figure()

# defining subplots
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

# plotting subplots
ax1.plot(numbers, numbers**5)
ax2.plot(numbers, numbers**2)
ax3.plot(numbers, numbers**3)
ax4.plot(numbers, numbers**4)
fig.suptitle("Numbers and it's Squares, Cubes, Quadraples")
plt.show()
```
